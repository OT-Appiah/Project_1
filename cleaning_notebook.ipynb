```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('bank_marketing.csv')
df.head()
```

```python
df.info()
```

```python
# cleaning data
df['job_clean'] = df['job'].str.replace('\.', '_', regex=True)
df['education_clean'] = df['education'].str.replace('\.', '_', regex=True)
df['education_clean'].replace('unknown', np.nan, inplace=True)
df['credit_default'] = df['credit_default'].astype(bool)
df['mortgage'] = df['mortgage'].astype(bool)
df['previous_outcome'] = df['previous_outcome'].astype(bool)
df['campaign_outcome'] = df['campaign_outcome'].astype(bool)
df['last_contact_date'] = pd.to_datetime(df['day'].astype(str) + '-' + df['month'].astype(str)  + '-2022', format='%d-%b-%Y')
df.head(10)
```

```python
#subsetting

client_df = df[['client_id', 'age', 'job_clean', 'marital', 'education_clean', 'credit_default', 'mortgage']]
client_df.rename(columns={'job_clean':'job', 'education_clean':'education'}, inplace=True)

client_df.head()
```

```python
campaign_df = df[['client_id', 'number_contacts', 'contact_duration', 'previous_campaign_contacts', 'previous_outcome', 'campaign_outcome', 'last_contact_date']]
campaign_df.head()
```

```python
economics_df = df[['client_id', 'cons_price_idx', 'euribor_three_months']]
economics_df.head()
```

```python
#save as csv
client_df.to_csv('client.csv', index=False)
campaign_df.to_csv('campaign.csv', index=False)
economics_df.to_csv('economics.csv', index=False)
```
